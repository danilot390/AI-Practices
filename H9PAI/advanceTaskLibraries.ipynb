## NumPy. Multi-Dimensional Array Operations
Create a 3D NumPy array representing a grayscale image, simulate applying a filter to it, and compute statistics for each chanel.
Requirements.
1. Create a random 3D array of shape (100, 100, 3) where each channel represents pixel intensities(calues between 0 and 255).
2. Normalize the array to have values between 0 and 1.
3. Apply a Gaussian blur (approximate with a kernel).
4. Compute the mean and standard deviation for each chanel.

```python
import numpy as np
from scipy.ndimage import gaussian_filter

# Create a random image with pixel values between 0 and 255
image = np.random.randint(0, 256, size=(100,100,3))

# Normalize the image to [0, 1]
normalized_image = image/255.0

# Apply Gaussian Blur
blurred_image = gaussian_filter(normalized_image, sigma=1)

# Compute mmean and standard deviation for each channel
means = blurred_image.mean(axis=(0, 1))
std_devs = blurred_image.std(axis=(0, 1))

print('Means per channel.', means)
print('Standard deviations per channel.', std_devs)
```

## Pandas. DataFrame Manipulating and Analysis
Analyze a dataset of AI research publications.
1. Simulate a dataset with columns: Year, Citations, and Authors.
2. Compute:
    - Average citations per year
    - The year with the most citations
    - The top 3 authors with the most total citations

```python
import pandas as pd

# Create a simulated dataset
data = {
    'Year': [2020, 2021, 2021, 2022, 2022, 2022],
    'Citations': [100, 150, 200, 300, 250, 400],
    'Authors': ['Alice', 'Bob', 'Alice', 'Charlie', 'Alice','Bob'],
}
df = pd.DataFrame(data)

# Compute average citations per year
avg_citations_per_year = df.groupby('Year')['Citations'].mean()
print('Average Citations Per Year')
print(avg_citations_per_year)

# Find the year with the most citations
most_cited_year = df.groupby('Year').sum().idxmax()
print(f'Year with the most citations: {most_cited_year}')

# Find the top 3 authors with the most total citations
top_authors = df.groupby('Authors')['Citations'].sum().sort_values(ascending=False).head(3)
print('Top 3 authors by Citations')
print(top_authors)
```

## Matplotlib. data Visualization
Visualize AI trends with multi-line plit and a histogram
1. Plot a multi-line chart showing the number of AI research publications and funding over the years.
2. Create a histogram of citation counts from the earlier pandas dataset.

```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt 

# Generate data using NumPy
models = [f'Model_{i}' for i in range(1,6)]
accuracy = np.random.uniform(0.8, 1.0, size=5)
precision = np.random.uniform(0.7, 1.0, size=5)
recall = np.random.uniform(0.6, 1.0, size=5)
training_time = np.random.randint(10, 50, size=5)

# Create the DataFrame
data = {
    'Model' : models,
    'Accuracy' : accuracy,
    'Precision' : precision,
    'Recall' : recall,
    'Training_time' : training_time,
}
df = pd.DataFrame(data)

# Analyze the data
avg_accuracy_per_model = df.groupby('Model')['Accuracy'].mean()
avg_precision_per_model = df.groupby('Model')['Precision'].mean()
avg_recall_per_model = df.groupby('Model')['Recall'].mean()
avg_training_per_model = df.groupby('Model')['Training_time'].mean()
print(f'Average Accuracy Per Model {avg_accuracy_per_model}')
print(f'Average Precision Per Model {avg_precision_per_model}')
print(f'Average Recall Per Model {avg_recall_per_model}')
print(f'Average Training time Per Model {avg_training_per_model}')

highest_accuracy_model = df.groupby('Model')['Accuracy'].sum().idxmax()
print(f'The highest Accuracy model -> {highest_accuracy_model}')

# Visualize the data
plt.figure(figsize=(8, 5))
plt.hist(df['Accuracy'], label='Accuracy', bins = 5, color='blue', edgecolor='black')
plt.hist(df['Precision'], label='Precision', bins = 5, color='red', edgecolor='black')
plt.title('Details of Models')
plt.xlabel('Model')
plt.ylabel('Accuracy/ Precision/ Recall/ Training time')
plt.grid()
plt.legend()
plt.show()

plt.figure(figsize=(8,5))
plt.title('Training time per Model')
plt.xlabel('Model')
plt.ylabel('Training time')
plt.legend()
plt.grid()

plt.plot(df['Model'], df['Training_time'], label='Training time', marker='#')

plt.show()

```
